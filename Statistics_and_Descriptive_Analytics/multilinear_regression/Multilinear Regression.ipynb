# Libraries, directory and data

```python
#libraries
import pandas as pd
import statsmodels.api as sm
import numpy as np
import seaborn as sns
```

```python
#load the data
df = pd.read_csv('salaries.csv')
df.head()
```

# Analysing the data

```python
#look at data set
df.head(1)
```

```python
#transform objects into dummies
df = pd.get_dummies(data=df, drop_first=True).astype(int)
df.head()
```

# Multilinear regression preparation

```python
#isolate X and Y
y = df.salary / 1000
X = df.drop(columns= ["salary", "yrs.service"])
X.head(1)
```

```python
#Add a constant
X = sm.add_constant(X)
X.head()
```

```python
#Training and Test Set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.2,
                                                    random_state=1502)
X_train.head(2)
```

# Multilinear Regression

```python
#Multilinear Regression
model = sm.OLS(y_train, X_train).fit()
print(model.summary())
```

# Accuracy Assessment

```python
#Predict
predictions = model.predict(X_test)
predictions[:4]
```

```python
#Accuracy assessment
from sklearn.metrics import mean_absolute_error, mean_squared_error
print("MAE:", round(mean_absolute_error(y_test, predictions),1))
print("RMSE:",round(np.sqrt(mean_squared_error(y_test, predictions)),1))
```
